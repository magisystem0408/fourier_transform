```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Nは周波数
N = 1024
# サンプリング周波数
fs = 100
# サンプリング周期
dt = 1 / fs
```

```python
# 時間定義
t = np.arange(0, N * dt, dt)

# 変動の周波数
f1 = 5
f2 = 10
f3 = 20

# 関数の重ね合わせ
y = np.sin(2 * np.pi * f1 * t) + 1.7 * np.sin(2 * np.pi * f2 * t) + 0.7 * np.sin(2 * np.pi * f3 * t) +0.5*np.random.randn(N)
```

```python
# 横軸時間
# 縦軸関数
plt.plot(t,y)

# 範囲を変更
plt.xlim(0,1)
```

```python
# FFTの実装
F =np.fft.fft(y)
```

```python
# 周波数定義
freq =np.fft.fftfreq(N,d =1/fs)
```

```python
# 振幅スペクトル、Fの絶対値を取ったもの
Amp =np.abs(F)
```

```python
# 意味のある周波数はサンプリング周波数の半分だけを使用する
plt.plot(freq[1:int(N/2)],Amp[1:int(N/2)])
plt.xlabel("Frequency[Hz]")
plt.ylabel("Amplitude[Hz]")
plt.title("FFT test")
# ピークが3つ
# これが、f1,f2,f3の値に対応するようになっている
```
